```python
import cv2
import matplotlib.pyplot as plt
```

```python
from collections import namedtuple #Para definir las tuplas OCRLocation
```

```python
def mostrarImagen(image):
    if len(image.shape)==3:
        img2 = image[:,:,::-1]
        plt.figure(figsize = (180,18))
        plt.imshow(img2,)
        
        plt.show()
    else:
        img2 = image
        plt.figure(figsize = (180,18))
        plt.imshow(img2,cmap='gray',shape=600)
        
        plt.show()
```

```python
image1 = cv2.imread("26032A00200143_page-0001.jpg")
image2 = cv2.imread("26032A00200143_page-0002.jpg")
mostrarImagen(image1)
mostrarImagen(image2)
```

```python
clone = image1.copy()
cv2.rectangle(clone, (940, 185), (1250, 225), (0, 255, 0), 2) #Termino municipal de
cv2.rectangle(clone, (380, 225), (680, 265), (0, 255, 0), 2) #Partido judicial de

mostrarImagen(clone)
```
